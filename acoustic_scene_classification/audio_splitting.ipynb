```python
import pandas as pd
import os
import concurrent.futures
import librosa
import time
import logging
```

```python
train_df = pd.read_pickle('/home/sauravpathak/acoustic_scene_classification_small/train_df.pkl')
test_df = pd.read_pickle('/home/sauravpathak/acoustic_scene_classification_small/test_df.pkl')
```

```python
print(train_df.iloc[0,2])
```

```python
train_df_indices = train_df.index.values
test_df_indices = test_df.index.values
```

```python
root_train_folder = '/home/sauravpathak/acoustic_scene_classification_small/Train_Split/'
root_test_folder = '/home/sauravpathak/acoustic_scene_classification_small/Test_Split/'
```

```python
os.cpu_count()
```

```python
def audio_split(idx, root_folder, df):
    
    audio, sr = librosa.load(df.iloc[idx,1], sr = 48000)
    buffer = 2 * sr
    samples_total = len(audio)
    samples_wrote = 0
    counter = 1
    while samples_wrote < samples_total:
        
        #check if the buffer is not exceeding total samples 
        if buffer > (samples_total - samples_wrote):
            buffer = samples_total - samples_wrote

        block = audio[samples_wrote : (samples_wrote + buffer)]
        out_filename = os.path.join(root_folder, df.iloc[idx,2], "split_" + str(counter) +  "_" + df.iloc[idx, 0])

        # Write 2 second segment
        librosa.output.write_wav(out_filename, block, sr)
        counter += 1
        samples_wrote += buffer
```

```python
def audio_split_train(idx):
    audio_split(idx, root_train_folder, train_df)
t1 = time.perf_counter()
with concurrent.futures.ThreadPoolExecutor() as executor:
    executor.map(audio_split_train, train_df_indices)
t2 = time.perf_counter()
print(f'Finished in {round(t2-t1, 2)} seconds')
```

```python
train_dir_folders = os.listdir('/home/sauravpathak/acoustic_scene_classification_small/Train_Split')
print(train_dir_folders)
train_dir_path = '/home/sauravpathak/acoustic_scene_classification_small/Train_Split'
```

```python
for folder in train_dir_folders:
    folder_path = os.path.join(train_dir_path, folder)
    folder_content = os.listdir(folder_path)
    for file in folder_content:
        file_name_list.append(file)
        file_path_list.append(os.path.join(folder_path, file))
        category_list.append(folder)
```

```python
def audio_split_test(idx):
    audio_split(idx, root_test_folder, test_df)
t1 = time.perf_counter()
with concurrent.futures.ThreadPoolExecutor() as executor:
    executor.map(audio_split_test, test_df_indices)
t2 = time.perf_counter()
print(f'Finished in {round(t2-t1, 2)} seconds')
```
